## Text-Preprocessing

```python
import re
import string
import nltk
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
text="""Hello !!! My name is Anjali. I am studing NLP in 2026.
        I love playing,played and plays football.
        This is a very interesting subject!!!"""
```

### 1. change text in lowercase

```python
text=text.lower()
print(text)
```

### 2.Remove Numbers

```python
text=re.sub(r'\d+'," ",text)
print("After removing Number : ",text)
```

### 3.Remove punctuation

```python
text=text.translate(str.maketrans("","",string.punctuation))
print("After removing the punctuation : ",text)
```

### 4.Tokenization

```python
tokens = word_tokenize(text)
print("After Tokenization : ",tokens)
```

### 5.Remove Stopwords

```python
from nltk.corpus import stopwords

stop_words=set(stopwords.words('english'))

filtered_words=[word for word in tokens if word not in stop_words]
print(filtered_words)
```

### 6.Stemming

```python
from nltk.stem import PorterStemmer,WordNetLemmatizer

ps=PorterStemmer()
stemmed_words=[ps.stem(word) for word in filtered_words]
print(stemmed_words)
```

### 7.Lemmatization

```python
lemmatizer= WordNetLemmatizer()
lemmatized_words=[lemmatizer.lemmatize(word,pos='v') for word in filtered_words]
print(lemmatized_words)
```
