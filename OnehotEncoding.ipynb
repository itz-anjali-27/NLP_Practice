## OnehotEncoding

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import pandas as pd
```

```python
# Sentence
sentence = "I love AI"
```

```python
# Convert sentence into list of words
words = sentence.split()
```

```python

# Convert into 2D array (required format)
words_array = np.array(words).reshape(-1, 1)

```

```python
# Create encoder
encoder = OneHotEncoder()
```

```python
# Fit and transform
encoded = encoder.fit_transform(words_array)
```

```python
# Print vocabulary
print("Vocabulary:", encoder.categories_)
```

```python
# Print encoded vectors
print(encoded)
```

```python
columns = encoder.get_feature_names_out(["word_array"])
print(columns)
```

```python
encoded_df = pd.DataFrame(encoded.toarray(), columns=columns)
print(encoded_df)
```

### with multiple columns

```python
data = {
    "City": ["Delhi", "Mumbai", "Delhi", "Chennai"],
    "Education": ["BCA", "BTech", "MCA", "BCA"],
    "Experience": [1, 2, 3, 1]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
encoded = encoder.fit_transform(df[["City", "Education"]])
```

```python
columns = encoder.get_feature_names_out(["City", "Education"])
```

```python
encoded_df = pd.DataFrame(encoded, columns=columns)
```

```python
print(encoded_df)
```

```python
# Combine with numeric column
final_df = pd.concat([encoded_df, df["Experience"]], axis=1)

```

```python
print(final_df)
```
