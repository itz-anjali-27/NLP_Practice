```python
#Bag of Words (BoW) is a simple and powerful technique in Natural Language Processing (NLP) used to convert text into numerical features
#so that machine learning models can understand it.
```

## Bag of words

```python
documents = [
    "I love NLP",
    "I love machine learning",
    "NLP is fun"
]
```

### Import Library

```python
from sklearn.feature_extraction.text import CountVectorizer
```

### Create Vectorizer Object

```python
vectorizer = CountVectorizer()
```

### Fit and Transform

```python
X = vectorizer.fit_transform(documents)
```

```python
print(vectorizer.vocabulary_)
```

### Convert to Array

```python
print(X.toarray())
```
